The aim of this project is to describe the dynamics of a population by using the Lotka-Volterra model. Two methods will be used to solve a set of ordinary differential equations (ODEs): Euler method and Runge-Kutta 4th order. 
The following is the set of equations to solve:


$\dot{x}=ax-bxy$ 



$\dot{y}=-cy+dxy$


* x is the population prey, y is the population of the predator 
* a: Natural growth rate of prey in the absence of predation
* b: Death rate due to predation
* c: Natural death rate of predators in the absence of prey
* d: Growth rate due to predation




Further info available in the pdf within this repo.

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
```

Initialization of the parameters of the model.

```python
alfa = 1

beta = 1

c= 1

d= 1
```

Euler method.

```python

def f(  x,y ): #VOLTERRA
    return alfa*x-beta*x*y

def g(x,y):
    return c*x*y-d*y

def euler_mthd( f, g , a, b, N, IV, V ):
 
    # determine step-size
    h = (b-a)/float(N)  
 
    # create mesh                         
    t = np.arange( a, b+h, h )
 
    # initialize w,z               
    w = np.zeros((N+1,)) #prey
    z = np.zeros((N+1,)) #predator

    # set initial values                     
    t[0], w[0] = IV   
    t[0], z[0] = V   
 
    # apply Euler's method                       
    for i in range(1,N+1):                       
        w[i] = w[i-1] + h * f( w[i-1], z[i-1])
        z[i] = z[i-1] + h * g( w[i-1], z[i-1])
    print("Euler's method")
    print("Number of Iterations: {} ".format (N))  
    figure(num=None, figsize=(12, 10), dpi=80, facecolor='w', edgecolor='k')

    plt.plot( t, w, label='Prey' )
    plt.plot( t, z, label='Predator' )
    plt.title( "Euler's Method Lotka-Volterra, N="+str(N) )
    plt.xlabel('time') 
    plt.ylabel('Population')
    plt.legend(loc=4)
    plt.grid()
    plt.show()
    #plt.savefig( 'euler_example_Lotaka-Volterra.png', fmt='PNG', dpi=100 )

    figure(num=None, figsize=(12, 10), dpi=80, facecolor='w', edgecolor='k')
    plt.plot( w, z, label='Prey-Predator' )
    plt.title( "Euler's Method Lotka-Volterra, N="+str(N) )
    plt.xlabel('Prey population') 
    plt.ylabel('Predator population')
    plt.legend(loc=4)
    plt.grid()
    plt.show( )
    #plt.savefig( 'euler_example_Lotaka-Volterra_phase_space.png', fmt='PNG', dpi=100 )

    return w
```

```python
euler_mthd( f, g , 0, 100, 100000, (0,2), (0,2)) 
```

Runge-Kutta IV


```python

def kutta(f, g, a, b, N, IV, V):

 

    #determine step-size
    step = (b-a)/float(N)
    # create mesh                         
    t = np.arange( a, b+step, step )

    x = np.arange(a, b, step)
    y = np.zeros(x.size)
    w = np.zeros(x.size)
    z = np.zeros(x.size)
    w[0] = IV
    y[0] = V  

    for i in range(x.size-1):
        k1 = f(w[i],y[i])
        k2 = f(w[i]+0.5*k1*step,y[i]+0.5*k1*step)
        k3 = f(w[i]+0.5*k2*step,y[i]+0.5*k2*step)
        k4 = f(w[i]+k3*step,y[i]+k3*step)
        w[i+1] = w[i]+(step/6.0)*(k1+2*k2+2*k3+k4)

        k1_2 = g(w[i], y[i])
        k2_2 = g(w[i]+0.5*k1_2*step, y[i]+0.5*k1_2*step)
        k3_2 = g(w[i]+0.5*k2_2*step,y[i]+0.5*k2*step)
        k4_2 = g(w[i]+k3_2*step,y[i]+k3*step)
        y[i+1] = y[i]+(step/6.0)*(k1_2+2*k2_2+2*k3_2+k4_2)

    print("Runge-Kutta")
    print("Evaluating ODE from {} to  {}".format(a,b))
    print("Number of points: {}".format(N))
    print("Step size {}".format(step))

    figure(num=None, figsize=(12, 10), dpi=80, facecolor='w', edgecolor='k')   
    plt.plot(x,w, label='Prey' )
    plt.plot(x,y, label='Predator' )
    plt.xlabel('time') 
    plt.ylabel('Population')
    plt.legend(loc=4)
    plt.grid()
    plt.show()
    #plt.savefig( 'runge_kutta_example_Lotaka-Volterra.png', fmt='PNG', dpi=100 )

    figure(num=None, figsize=(12, 10), dpi=80, facecolor='w', edgecolor='k')
    plt.plot( w, y, label='Prey-Predator' )
    plt.title( "Runge-Kuttas's Method Lotka-Volterra, N="+str(N) )
    plt.xlabel('Prey population') 
    plt.ylabel('Predator population')
    plt.legend(loc=4)
    plt.grid()
    plt.show()
    #plt.savefig( 'runge_kutta_example_Lotaka-Volterra_phase_space.png', fmt='PNG', dpi=100 )
```

```python
# Solving a Sytem of ODEs with TensoFlow

# Ideas and ispiration taken from TensorFlow Cookbook! Thank you. 

import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.python.framework import ops
from matplotlib.pyplot import figure



def tensor_ode(N, x0, y0):
    ops.reset_default_graph()
    # Open interactive graph session
    sess = tf.Session()
    
    x_initial = tf.constant(x0)
    y_initial = tf.constant(y0)
    X_t1 = tf.Variable(x_initial)
    Y_t1 = tf.Variable(y_initial)

    # Make the placeholders
    t_delta = tf.placeholder(tf.float32, shape=())
    a = tf.placeholder(tf.float32, shape=())
    b = tf.placeholder(tf.float32, shape=())
    c = tf.placeholder(tf.float32, shape=())
    d = tf.placeholder(tf.float32, shape=())
    
    
    X_t2 = X_t1 + (a * X_t1 + b * X_t1 * Y_t1) * t_delta
    Y_t2 = Y_t1 + (c * Y_t1 + d * X_t1 * Y_t1) * t_delta
    # Update to New Population
    step = tf.group(
        X_t1.assign(X_t2),
        Y_t1.assign(Y_t2))
    init = tf.global_variables_initializer()
    sess.run(init)
    # Run the ODE
    prey_values = []
    predator_values = []
    for i in range(N):
        # Step simulation (using constants for a known cyclic solution)
        step.run({a: (1.0), b: (-1.0), c: -1.0, d: 1.0, t_delta: 0.001}, session=sess)
        # Store each outcome
        temp_prey, temp_pred = sess.run([X_t1, Y_t1])
        prey_values.append(temp_prey)
        predator_values.append(temp_pred)
        
    print("Tensorflow solving ODE")
    print("Number of points: {}".format(N))
    figure(num=None, figsize=(12, 10), dpi=80, facecolor='w', edgecolor='k')   
    plt.plot(prey_values)
    plt.plot(predator_values)
    plt.xlabel('time (arbitrary units)')
    plt.ylabel('Population')
    plt.legend(['Prey', 'Predator'], loc='lower left')
    plt.show()
    figure(num=None, figsize=(12, 10), dpi=80, facecolor='w', edgecolor='k')   
    plt.plot(prey_values, predator_values)
    plt.xlabel('Prey population')
    plt.ylabel('Predators population')
    plt.show()
```

```python
tensor_ode(100000, 4.0,1.0)
```
